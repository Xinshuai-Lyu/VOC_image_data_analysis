```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob, os
import seaborn as sns
```

### Num of images and objects

```python
images_metadata_dataframe = pd.read_csv("./images_metadata.csv")
```

```python
num_images = len(np.unique(images_metadata_dataframe["filename"]))
num_objects = len(images_metadata_dataframe)
df = pd.DataFrame({"x":["num_of_images", "num_of_objects"],
                   "num":[num_images, num_objects]})
ax = sns.barplot(x="x", y="num", data=df)
ax.text(0, num_images, num_images, color='black', ha="center")
ax.text(1, num_objects, num_objects, color='black', ha="center")
plt.show()
```

### How many objects an image contains?

```python
objects_count = images_metadata_dataframe.groupby(["filename"]).agg(["count"])["class"]
```

```python
objects_count.describe()
```

```python
sns.boxplot(data=objects_count["count"], orient="h")
plt.show()
```

```python
objects_count_describe = objects_count.describe()["count"]
v25 = objects_count_describe[4]
v75 = objects_count_describe[6]
iqr = 1.5 * (v75 - v25)
min_objects_in_acceptable_range = v25 - iqr
max_objects_in_acceptable_range = v75 + iqr
number_of_outliers = sum((objects_count["count"] < min_objects_in_acceptable_range) | (objects_count["count"] > max_objects_in_acceptable_range))
outliers_percentage = number_of_outliers / num_images
print(f"{max_objects_in_acceptable_range} is the max normal objects an image contains")
print(f"boxplot number of outliers:{number_of_outliers}, outliers percentage:{outliers_percentage}")
```

### missing values check

```python
attributes_missing_value_num = np.sum(images_metadata_dataframe.isnull())
attributes_missing_value_index =attributes_missing_value_num.index
df = pd.DataFrame({"x":attributes_missing_value_index,
                   "missing_num":attributes_missing_value_num})
plt.figure(figsize=(18, 2))
ax = sns.barplot(x="x", y="missing_num", data=df)
ax.set(ylim=(0, 0.1))
plt.show()
```

### Objects distribution

```python
objects = images_metadata_dataframe["class"].value_counts()
objects_percentages = round(objects / np.sum(objects), 2)
objects_indexs = objects.index
df = pd.DataFrame({"x":objects_indexs,
                   "percentage":objects_percentages})
plt.figure(figsize=(18, 2))
ax = sns.barplot(x="x", y="percentage", data=df)
for i in range(len(objects)):
    ax.text(i, objects_percentages[i], objects_percentages[i], color='black', ha="center")
plt.show()
```
