```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import tensorflow as tf

import cv2
```

```python
epsilon = tf.keras.backend.epsilon()

def xywhs_to_bbs(xywhs):
    """ Transform xywhs to bounding boxes xmin,ymin,xmax,ymax

        parameters:
        xywhs list: in which each record is x, y, w, h

        return:
        bbs list: in which each record is xmin,ymin,xmax,ymax
    """
    xmin = xywhs[..., 0] - xywhs[..., 2]/2
    ymin = xywhs[..., 1] - xywhs[..., 3]/2
    xmax = xywhs[..., 0] + xywhs[..., 2]/2
    ymax = xywhs[..., 1] + xywhs[..., 3]/2
    return np.stack([xmin, ymin, xmax, ymax], axis=-1)

def compute_iou(y_true_box, y_pred_box):
    center_x1, center_y1, box_width1, box_height1 = y_true_box
    center_x2, center_y2, box_width2, box_height2 = y_pred_box
    x1_min, x1_max = center_x1 - box_width1 / 2.0, center_x1 + box_width1 / 2.0
    y1_min, y1_max = center_y1 - box_height1 / 2.0, center_y1 + box_height1 / 2.0
    x2_min, x2_max = center_x2 - box_width2 / 2.0, center_x2 + box_width2 / 2.0
    y2_min, y2_max = center_y2 - box_height2 / 2.0, center_y2 + box_height2 / 2.0
    # study https://www.pyimagesearch.com/2016/11/07/intersection-over-union-iou-for-object-detection/
    xA = tf.math.maximum(x1_min, x2_min)
    yA = tf.math.maximum(y1_min, y2_min)
    xB = tf.math.minimum(x1_max, x2_max)
    yB = tf.math.minimum(y1_max, y2_max)
    interArea = tf.math.maximum(0.0, xB - xA) * tf.math.maximum(0.0, yB - yA)
    boxAArea = box_width1 * box_height1
    boxBArea = box_width2 * box_height2
    all_area = boxAArea + boxBArea - interArea + epsilon
    iou = interArea / all_area
    return iou
```

### Data preparation

```python
images_metadata_dataframe = pd.read_csv("./images_metadata.csv")
x_y_w_h_filename = images_metadata_dataframe[["x", "y", "w", "h", "filename"]]
iou_threshold = 0.5
images_classes_iou_greater_than_point_5 = []
unique_filenames = x_y_w_h_filename["filename"].unique()
```

### Check and record the image which has iou conflict problem

```python
for filename in unique_filenames:
    display_column = ["x", "y", "w", "h"]
    xywhs = x_y_w_h_filename[display_column][x_y_w_h_filename["filename"] == filename]
    xywhs = xywhs.to_numpy()
    i = 0
    while i < len(xywhs):
        # Searched one IOU conflict in an image
        can_break = False
        j = i + 1
        while j < len(xywhs):
            if compute_iou(xywhs[i], xywhs[j]) >= iou_threshold:
                images_classes_iou_greater_than_point_5.append(filename)
                can_break = True
            if can_break: break
            j += 1
        if can_break: break
        i += 1
```

### Findings

    - 745 (4%) images have iou conflict problems
    

```python
print(f"{len(images_classes_iou_greater_than_point_5)} images have iou conflict")
print(f"{round(len(images_classes_iou_greater_than_point_5) / len(unique_filenames) * 100, 2)}% images have iou conflict")
```

### Check the top 5 classes which makes the iou conflict problem happen

```python
all_iou_conflits_objects = []
unique_filenames = images_classes_iou_greater_than_point_5
for filename in unique_filenames:
    display_column = ["x", "y", "w", "h", "class"]
    xywhs_classes = images_metadata_dataframe[display_column][x_y_w_h_filename["filename"] == filename]
    xywhs_classes = xywhs_classes.to_numpy()
    xywhs = xywhs_classes[..., 0:4]
    classes = xywhs_classes[..., -1]
    i = 0
    while i < len(xywhs):
        # Searched all IOU conflicts in an image
        j = i + 1
        while j < len(xywhs):
            if compute_iou(xywhs[i], xywhs[j]) >= iou_threshold:
                # These two objects cause the iou conflict problem
                all_iou_conflits_objects.append(classes[i])
                all_iou_conflits_objects.append(classes[j])
            j += 1
        i += 1
```

### Data preparation

```python
df = pd.DataFrame({"all_iou_conflits_objects":all_iou_conflits_objects})
iou_conflits_objects_counts = df["all_iou_conflits_objects"].value_counts()
df = pd.DataFrame({"x":iou_conflits_objects_counts.index,
                   "count":iou_conflits_objects_counts})
```

### Visualize the top 5 iou conflict classes
    - person, chair, sofa, motorbike, bicycle

```python
plt.figure(figsize=(18, 5))
ax = sns.barplot(x="x", y="count", data=df)
for i in range(len(iou_conflits_objects_counts)):
    ax.text(i, iou_conflits_objects_counts[i], iou_conflits_objects_counts[i], color='black', ha="center")
plt.show()
```
